```python
import pandas as pd
    #Create Series
planets = pd.Series(["Mercury","Venus","Earth","Mars","Jupiter","Saturn","Uranus","Neptune","Pluto"])
moons = pd.Series([0,0,1,2,95,146,27,14,5])
planet_diameter_Km = pd.Series([4890,12104,12756,6794,142984,120536.51118,2378.6])
planet_type = pd.Series(["Terrestrial","Terrestrial","Terrestrial","Terrestrial","Gas","Gas",
                         "Ice","Ice","Dwarf"])
dist_frm_sun = pd.Series([58,108,150,227.9,778,1400,2900,4500,5900])
    #Create Dataframe
Planets_df = pd.DataFrame({"Planets":planets,"Moons":moons,"Planet Diameter(Km)":planet_diameter_Km,
                           "Planet Type":planet_type,"Dist frm sun":dist_frm_sun})
Planets_df
```
